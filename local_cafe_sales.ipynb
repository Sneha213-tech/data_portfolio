```python
copy_cafe = cafe.copy()
```

```python
copy_cafe["Transaction ID"] = pd.to_numeric(cafe["Transaction ID"].str.replace("TXN_", "", regex=False), errors="coerce")
copy_cafe['Transaction ID']
```

```python
copy_cafe[["Item", "Payment Method", "Location"]] = cafe[["Item", "Payment Method", "Location"]].astype('string')
```

```python
copy_cafe[["Quantity", "Price Per Unit", "Total Spent"]] = cafe[["Quantity", "Price Per Unit", "Total Spent"]].apply(pd.to_numeric, errors="coerce")
```

```python
copy_cafe['Transaction Date'] = pd.to_datetime(copy_cafe['Transaction Date'], errors='coerce')
```

```python
copy_cafe.info()
```

```python
for column in cafe.columns:
    print(f"\n{column}")
    print(f"NAN values: {cafe[column].isna().sum()}")
    print(f"ERROR values: {(cafe[column] == 'ERROR').sum()}")
    print(f"UNKNOWN values: {(cafe[column] == 'UNKNOWN').sum()}")
```

```python
copy_cafe.replace(['ERROR','UNKWOWN'],np.nan , inplace=True)
```
